#European Life-Work Balance Index
Información de datos extraido de Kaggle: https://www.kaggle.com/datasets/yigitnewday/european-life-work-balance-index

#Preguntas que buscamos responder con este conjunto

* ¿Qué países tienen el mejor balance ?
* ¿Cuáles paises obtienen mayor salario?  
* ¿Qué paises tienen una mayor aceptacion a la inclusividad LGBT?  
* ¿Cuásntas cantidad de horas trabaja cada pais por semana?
* ¿Cuáles son los paises con mas alto porcentaje de felicidad?

En nuestro proyecto encontraremos estos Datos:

1. Overall Rank : Ranking sobre las mejores posiciones en los países europeos sobre la calidad de vida-trabajo.
2. highest_paying_countries : Los salarios mas altos de cada uno.
3. Percent of Total Hours Worked Porcentaje de horas de trabajo semanal por país.
4. LGBTQ+ Inclusivity: Paises con mas alta inclusividad LGBT.
5. happiest_countries : Felicidad en porcentaje de países europeos.

```
import pandas as pd
```

```
data =pd.read_csv('//content/life_work_balance_europe.csv')
data.head()
```

```
data.info()
```

```
data.describe()
```

```
# Ordenamos el DataFrame por 'Overall Rank' de manera ascendente para obtener los mejores rangos primero.
ranked_data = data.sort_values(by='Overall Rank', ascending=True)
```

```
# Para visualizar los países con el más alto 'Overall Rank', simplemente mostramos las primeras filas.
print(ranked_data[['Overall Rank', 'European Country']].head())
```

```
# Ordenamos el DataFrame por 'Minimum Wage(USD/hr)' de manera descendente para obtener los salarios más altos primero.
highest_paying_countries = data.sort_values(by='Minimum Wage(USD/hr)', ascending=False)
```

```
# Ahora, visualizamos los países junto con sus salarios mínimos.
print(highest_paying_countries[['European Country', 'Minimum Wage(USD/hr)']].head())
```

```
#Ordenamos el DataFrame por 'LGBTQ+ Inclusivity' de manera descendente.
inclusive_countries = data.sort_values(by='LGBTQ+ Inclusivity', ascending=False)
```

```
# Para visualizar los países junto con su grado de integridad LGBTQ+, imprimimos los primeros registros.
print(inclusive_countries[['European Country', 'LGBTQ+ Inclusivity']].head())
```

```
# Calculamos el total de horas trabajadas por todos los países.
total_hours_worked = data['Average Hours Worked(per week)'].sum()
```

```
# Creamos una nueva columna en el DataFrame para los porcentajes.
data['Percent of Total Hours Worked'] = (data['Average Hours Worked(per week)'] / total_hours_worked) * 100
```

```
#Ordenamos el DataFrame por 'Percent of Total Hours Worked' de manera descendente si queremos ver primero los países con mayor porcentaje.
data_sorted_by_percentage = data.sort_values(by='Percent of Total Hours Worked', ascending=False)
```

```
# Para visualizar los países junto con su porcentaje del total de horas trabajadas, imprimimos los registros.
print(data_sorted_by_percentage[['European Country', 'Average Hours Worked(per week)', 'Percent of Total Hours Worked']])
```

```
# Ordenamos el DataFrame por 'Happiness Index(out of 10)' de manera descendente.
happiest_countries = data.sort_values(by='Happiness Index(out of 10)', ascending=False)
```

```
# Para visualizar los países junto con su índice de felicidad, imprimimos los primeros registros.
print(happiest_countries[['European Country', 'Happiness Index(out of 10)']].head())
```

```
 #Ordenamos el DataFrame por 'Happiness Index(out of 10)' de manera descendente para ver los países más felices.
happiest_countries = data.sort_values(by='Happiness Index(out of 10)', ascending=False)
```

```
# Mostramos todos los países junto con su índice de felicidad.
print(happiest_countries[['European Country', 'Happiness Index(out of 10)']])
```

```
#Gráfico de barras
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Ordenamos el DataFrame por 'Happiness Index(out of 10)' de manera descendente para ver los países más felices.
happiest_countries = data.sort_values(by='Happiness Index(out of 10)', ascending=False)
```

```
# Crear el gráfico de barras
plt.figure(figsize=(15, 10))
plt.barh(happiest_countries['European Country'], happiest_countries['Happiness Index(out of 10)'], color='skyblue')
plt.xlabel('Happiness Index(out of 10)')
plt.title('Happiness Index of Countries')
plt.gca().invert_yaxis()  # Invertir el eje y para que el país con el índice más alto esté en la parte superior
plt.show()
```
